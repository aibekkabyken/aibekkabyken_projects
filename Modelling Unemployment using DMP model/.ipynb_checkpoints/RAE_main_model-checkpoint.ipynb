```python
'''These are the parameters values for calibration'''
# rho - 
p = 0.1360085077925485

#lambda -
l = 0.056

# beta
b = 0.5

# risk-free rate
r = 0.02

# productivity of domesetic
yd = 1
bd = 0.15

# productivity of foreign and unemployment benefits
yf = 1.2529403713122165
bf = 0.1

#visa fees - fixed cost, cost per employee, percentage that foreigns absorb
tau = 0.13
pi = 0.7

#sponsorship cost for foreign and domestic
bnd = (1-pi)*tau
bnf = pi*tau

# cost of vacancy
k = 0.5

#mu
mu = 0.5

# original fraction of non-EU to UK + EU grads, from HESA Graduate Outcomes 2018/2019
# dictionary below is for the total response amount and response rate respectively
res = {
    'UK' : [190065, 0.55],
    'EU' : [13315, 0.5],
    'Non-EU' : [38420, 0.33]
}
c = (res['Non-EU'][0]*(1/res['Non-EU'][1]))/(res['UK'][0]*(1/res['UK'][1])+res['EU'][0]*(1/res['EU'][1]))
print (c)
```

```python
import scipy.stats as st
import pandas as pd
import numpy as np

# imports and reads the cleaned dataset
wages = pd.read_csv("wages.csv")

avg = wages['wages'].mean()
std = wages['wages'].std()

# this function finds the mean of yf above a given wf
def find_yf(min_wage, average):
    total = 0
    track = 0
    min_yf = wages.loc[wages['wages'] >= min_wage]
    avg_min_yf = min_yf['wages'].mean()
    productivity = (avg_min_yf)/average
    return productivity
    
# this function finds the cdf for rho above a given wf
def find_cdf(min_wage, average, standard, initial):
    min_yp = wages.loc[wages['wages'] >= min_wage]
    cdf = initial*(len(min_yp)/len(wages))
    return cdf
```

```python
'''This function here returns the yf and rho for a  wf'''
def find_minw(begin, end, interval, is_yf):
    yf = []
    p = []
    count = 0
    if begin == end:
        minww = [begin]
        yf.append(find_yf(minww[count], avg))
        p.append(find_cdf(minww[count], avg, std, c))
        if is_yf == True:
            return yf
        else:
            return p
    else:
        minww = np.arange(start=begin, stop=end, step=interval) 
        while count < len(minww):
            if is_yf == True:
                yf.append(find_yf(minww[count], avg))
            else:
                p.append(find_cdf(minww[count], avg, std, c))
            count += 1
        if is_yf == True:
            return yf
        else:
            return p

# over here we determined our yf and rho that we use in our model
yfn = find_minw(25600, 25600, 500, True)
pn = find_minw(25600, 25600, 500, False)
print(pn)
print(yfn)
```

```python
import sympy as sym

xaxis = []
gamma = []
ths = []

# this function scatters for a suitable root of gamma and ths so that 0>gamma>1 and theta>0
def find_root(z):
    count = 0
    count1 = 0
    final = []
    while count < len(z):
        while count1 < len(z[count]):
            num = z[count][count1]
            if num > 0:
                count1 += 1
            else:
                count1 = 3
            if count1 == 2:
                final.append(z[count][0])
                final.append(z[count][1])
                return final
        count1 = 0
        count += 1

# this functions essentially solves our system of equations using equations (30) and (31)
def long_solution(start, end, interval, is_minw):
    count = 0
    minws = start
    until = (end-start)/interval
    if start == end:
        until = 1
    while count < until:
        xaxis.append(start)
        if is_minw == False:
            bnd = (1-pi)*(start)
            bnf = pi*(start)
            global yf
            global p
        else: 
            yfw = find_minw(minws, end, interval, True)
            pw = find_minw(minws, end, interval, False)
            yf = float(round(yfw[count], 5))
            p = float(round(pw[count], 5))
            bnd = (1-pi)*tau
            bnf = pi*tau
        p1 = float(round((yf-bnf-bf), 5))
        p2 = float(round((yd-bnd-bd), 5))
        rmu = 1-mu
        g,t = sym.symbols('g, t')
        # below is equation (30)
        f = sym.Eq(((1-g)*bnd/(r+l)), g*p1/(r+l+(t**rmu)*b*((t**rmu)*(1-g)*p-l*(p-g))/((t**rmu)*g*(1-p))))
        # below is equation (31)
        h = sym.Eq(((1-b)*(g*p1+(1-g)*p2))/(r+l+(t**rmu)*b), k*(t**(mu)))
        roots = sym.solve([f,h],(g,t), real=True)
        z = find_root(roots)
        gamma.append(float(z[0]))
        ths.append(float(z[1]))
        start += interval
        count += 1

# very crucial that you wait until the function prints done, as it takes time usually 2 mins to process
is_minw = False
minw_range = [0.07, 0.30, 0.01, is_minw]
long_solution(minw_range[0], minw_range[1], minw_range[2], is_minw)
print ("done")
```

```python
'''This section here adds our other values using gamma and theta'''
qths = []
phi = []
uf = []
uf_unweighted = []
ud = []
u = []

count = 0
while count < len(gamma):
    if is_minw == True:
        pw = find_minw(minw_range[0], minw_range[1], minw_range[2], False)
        p = pw[count]
    val_qths = ths[count]**(1-mu)
    qths.append(val_qths)
    val_p = ((qths[count]*(1-gamma[count])*p + (l*(p-gamma[count])))/(qths[count]*gamma[count]*(1-p)))
    if val_p < 0:
        val_p = 0
    if val_p > 1:
        val_p = 1
    phi.append(val_p)
    val_u = (l*(1-p))/((qths[count]+l)*(1-gamma[count]))
    u.append(val_u)
    val_ud = (1-gamma[count])*val_u/(1-p) 
    ud.append(val_ud)
    uf_unchecked = (l/(qths[count]*val_p+l))
    uf_unweighted.append(uf_unchecked)
    count += 1
```

```python
''' Is not used in the paper, but this finds the derivative of a given variable'''
dxaxis = []
for i in xaxis:
    dxaxis.append(i)
dxaxis.pop(0)

def derivative(array):
    result = []
    count = 0
    while count < (len(array)-1):
        if array[count] == 0:
            array[count] = 0.000001
        num = ((array[count+1] - array[count])/array[count])
        result.append(num)
        count += 1
    return result
```

```python
# repeats same but for wf
qths = []
phi = []
uf = []
uf_unweighted = []
ud = []
u = []

count = 0
while count < len(gamma):
    if is_minw == True:
        pw = find_minw(minw_range[0], minw_range[1], minw_range[2], False)
        p = pw[count]
    val_qths = ths[count]**(1-mu)
    qths.append(val_qths)
    val_p = ((qths[count]*(1-gamma[count])*p + (l*(p-gamma[count])))/(qths[count]*gamma[count]*(1-p)))
    if val_p < 0:
        val_p = 0
    if val_p > 1:
        val_p = 1
    phi.append(val_p)
    val_u = (l*(1-p))/((qths[count]+l)*(1-gamma[count]))
    u.append(val_u)
    val_ud = (1-gamma[count])*val_u/(1-p) 
    ud.append(val_ud)
    uf_unchecked = (l/(qths[count]*val_p+l))
    uf_unweighted.append(uf_unchecked)
    count += 1
```

```python
'''In this section we find the wages in our model using equations (34) and (35)'''
yfw = find_minw(minw_range[0], minw_range[1], minw_range[2], True)
pw = find_minw(minw_range[0], minw_range[1], minw_range[2], False)
def find_model_wage(is_domestic):
    wages = []
    count = 0
    while count < len(qths):
        bot = qths[count]*b
        if is_domestic == False:
            salary = b*(yfw[count]-bnf)+((1-b)*(bf*(r+l)+bot*phi[count]*(yfw[count]-bnf))/(r+l+bot*phi[count]))
        else:
            salary = b*(yd-phi[count]*bnd)+((1-b)*(bd*(r+l)+bot*(yd-phi[count]*bnf))/(r+l+bot))
        wages.append(salary)
        count += 1
    return wages

wd = find_model_wage(True)
wi = find_model_wage(False)
```

```python
# this finds our average productivity based on equation (33)
total_y = []
pw = find_minw(minw_range[0], minw_range[1], minw_range[2], False)
yfw = find_minw(minw_range[0], minw_range[1], minw_range[2], True)

count = 0
while count < len(pw):
    total_y.append(yd * (1-pw[count]) + yfw[count] * pw[count])
    count += 1
```
